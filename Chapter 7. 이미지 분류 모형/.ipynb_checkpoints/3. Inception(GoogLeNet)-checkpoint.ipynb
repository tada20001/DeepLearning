ì´ ì¤‘, 1x1 convolution ì—°ì‚°ì€ ëª¨í˜¸í•˜ê²Œ ì—¬ê²¨ì§ˆ ê²ƒì´ë‹¤. í•˜ì§€ë§Œ, 1x1 convolutionì€ Inception moduleì—ì„œ í•µì‹¬ ì—­í• ì„ í•œë‹¤. 1x1 convolution ì—°ì‚°ì˜ ê¸°ëŠ¥ì€ ë‘ê°€ì§€ ì´ë‹¤.

* **ì²«ë²ˆì§¸ëŠ” ì±„ë„ì˜ ìˆ˜ë¥¼ ì¡°ì ˆí•˜ëŠ” ê¸°ëŠ¥**ì´ë‹¤. ì±„ë„ ìˆ˜ë¥¼ ì¡°ì •í•œë‹¤ëŠ” ì˜ë¯¸ëŠ” ë‹¤ì‹œ ë§í•´, **ì±„ë„ ê°„ì˜ Correlationì„ ì—°ì‚°í•œë‹¤**ëŠ” ì˜ë¯¸ë¼ê³  í•  ìˆ˜ ìˆë‹¤. ê¸°ì¡´ì˜ convolution ì—°ì‚°ì€, ì˜ˆë¥¼ ë“¤ì–´ 3x3ì˜ ì»¤ë„ì„ ì´ìš©í•´ ì—°ì‚°ì„ í•  ê²½ìš°, 3x3 í¬ê¸°ì˜ ì§€ì—­ ì •ë³´ì™€ í•¨ê»˜ ì±„ë„ ê°„ì˜ ì •ë³´ ë˜í•œ ê°™ì´ ê³ ë ¤í•˜ì—¬ í•˜ë‚˜ì˜ ê°’ìœ¼ë¡œ ë‚˜íƒ€ë‚¸ë‹¤. ë‹¤ì‹œ ë§í•˜ë©´, í•˜ë‚˜ì˜ ì»¤ë„ì´ 2ê°€ì§€ì˜ ì—­í• ì„ ëª¨ë‘ ìˆ˜í–‰í•´ì•¼ í•œë‹¤ëŠ” ê²ƒì´ë‹¤. ëŒ€ì‹  ì´ì „ì— 1x1 convolutionì„ ì‚¬ìš©í•œë‹¤ë©´, 1x1ì€ ì±„ë„ì„ ì¡°ì ˆí•˜ëŠ” ì—­í• ì„ í•˜ê¸° ë•Œë¬¸ì—, ìµœì í™” ê³¼ì •ì—ì„œ ì±„ë„ ê°„ì˜ íŠ¹ì§•ì„ ì¶”ì¶œ í•  ê²ƒì´ê³ , 3x3ì€ ì´ë¯¸ì§€ì˜ ì§€ì—­ì •ë³´ì—ë§Œ ì§‘ì¤‘í•˜ì—¬ íŠ¹ì§•ì„ ì¶”ì¶œí•˜ë ¤ í•  ê²ƒì´ë‹¤. **ì—­í• ì„ ì„¸ë¶„í™” í•´ì¤€ ê²ƒ**ì´ë‹¤. ì±„ë„ê°„ì˜ ê´€ê³„ì •ë³´ëŠ” 1x1 convolutionì— ì‚¬ìš©ë˜ëŠ” íŒŒë¼ë¯¸í„°ë“¤ ë¼ë¦¬, ì´ë¯¸ì§€ì˜ ì§€ì—­ ì •ë³´ëŠ” 3x3 convolutionì— ì‚¬ìš©ë˜ëŠ” íŒŒë¼ë¯¸í„°ë“¤ ë¼ë¦¬ ì—°ê²°ëœë‹¤ëŠ” ì ì—ì„œ ë…¸ë“œ ê°„ì˜ ì—°ê²°ì„ ì¤„ì˜€ë‹¤ê³  ë³¼ ìˆ˜ ìˆë‹¤.

* ê·¸ë¦¬ê³  ë‘ë²ˆì§¸ëŠ” **1x1 convolution ì—°ì‚°ìœ¼ë¡œ ì´ë¯¸ì§€ì˜ ì±„ë„ì„ ì¤„ì—¬ì¤€ë‹¤**ë©´, 3x3ê³¼ 5x5 convolution ë ˆì´ì–´ì—ì„œì˜ íŒŒë¼ë¯¸í„° ê°œìˆ˜ë¥¼ ì ˆì•½í•  ìˆ˜ ìˆë‹¤. ì´ ë•ë¶„ì— ë§ì„ ê¸°ì¡´ì˜ CNN êµ¬ì¡°ë“¤ ë³´ë‹¤ ë”ìš± ê¹Šê²Œ ë§Œë“¤ê³ ë„ íŒŒë¼ë¯¸í„°ê°€ ê·¸ë¦¬ í¬ì§€ ì•Šë‹¤.

ì•„ë˜ì˜ ì½”ë“œëŠ” Inception moduleì„ êµ¬í˜„í•œ ê²ƒ ì´ë‹¤.

```python
import tensorflow as tf
tf.logging.set_verbosity(tf.logging.ERROR)
from keras import layers

# Inception ëª¨ë“ˆ ì •ì˜
def inception_module(x, o_1=64, r_3=64, o_3=128, r_5=16, o_5=32, pool=32):
    """
    # Augments
    x : ì…ë ¥ì´ë¯¸ì§€
    o_1 : 1 x 1 convolution ì—°ì‚° ì¶œë ¥ê°’ì˜ ì±„ë„ ìˆ˜
    r_3 : 3 x 3 convolution ì´ì „ì— ìˆëŠ” 1x1 convolution ì¶œë ¥ê°’ì˜ ì±„ë„ ìˆ˜
    o_3 : 3 x 3 convolution ì—°ì‚° ì¶œë ¥ê°’ì˜ ì±„ë„ ìˆ˜
    r_5 : 5 x 5 convolution ì´ì „ì— ìˆëŠ” 1x1 convolution ì¶œë ¥ê°’ì˜ ì±„ë„ ìˆ˜
    o_5 : 5 x 5 convolution ì—°ì‚° ì¶œë ¥ê°’ì˜ ì±„ë„ ìˆ˜
    pool : maxplooling ë‹¤ìŒì˜ 1x1 convolutin ì¶œë ¥ê°’ì˜ ì±„ë„ ìˆ˜
    
    # returns : 4
    4 ì¢…ë¥˜ ì—°ì‚° ê²°ê³¼ ê°’ì„ ì±„ë„ ë°©í–¥ìœ¼ë¡œ í•©ì¹œ ê²°ê³¼
    """
    
    x_1 = layers.Conv2D(o_1, 1, padding='same')(x)
    
    x_2 = layers.Conv2D(r_3, 1, padding='same')(x)
    x_2 = layers.Conv2D(o_3, 3, padding='same')(x_2)
    
    x_3 = layers.Conv2D(r_5, 1, padding='same')(x)
    x_3 = layers.Conv2D(o_5, 5, padding='same')(x_3)
    
    x_4 = layers.MaxPooling2D(pool_size=(3, 3), strides=1, padding='same')(x)
    x_4 = layers.Conv2D(pool, 1, padding='same')(x_4)
    
    return layers.concatenate([x_1, x_2, x_3, x_4])
```

## GoogLeNetì˜ êµ¬ì¡°

### GoogLeNet

ìœ„ í‘œì—ì„œ inception module ì—´ì˜ ì˜ë¯¸ëŠ”, ì˜ˆë¥¼ ë“¤ì–´ "64,(96,128),(16,32),32" ëŠ” inception module ê·¸ë¦¼ì—ì„œ ê°€ì¥ ì™¼ìª½ì˜ 1x1 convolution ì—°ì‚°ì˜ ì¶œë ¥ í”¼ì³ë§µì˜ ê°¯ìˆ˜ê°€ 64, ë‘ë²ˆì§¸ 1x1, 3x3 convolution ì—°ì‚°ì˜ ì¶œë ¥ í”¼ì³ë§µì˜ ê°¯ìˆ˜ëŠ” ê°ê° 96,128, ì„¸ë²ˆì§¸ 1x1, 5x5 convolution ì—°ì‚°ì˜ ì¶œë ¥ í”¼ì³ë§µì˜ ê°¯ìˆ˜ëŠ” ê°ê° 16,32 ë§ˆì§€ë§‰ maxpooling í›„ì˜ 1x1 convolution ì—°ì‚°ì˜ ì¶œë ¥ í”¼ì³ë§µì˜ ê°¯ìˆ˜ëŠ” 32 ë€ ì˜ë¯¸ ì´ë‹¤.

ëª¨ë¸ì˜ ì²«ë²ˆì§¸ ë ˆì´ì–´ë¥¼ ë³´ë©´, ì¼ë°˜ì ì¸ CNN ëª¨ë¸ê³¼ í¬ê²Œ ë‹¤ë¥´ì§€ ì•Šë‹¤. 11x11ì˜ ì»¤ë„ ì‚¬ì´ì¦ˆì™€ 4ì˜ stridesë¡œ convolution ì—°ì‚°ì„ í•œë‹¤. ì´ ì²«ë²ˆì§¸ ë ˆì´ì–´ë¥¼ Stem layerë¼ê³  ë¶€ë¥¸ë‹¤. ì´ ì²«ë²ˆì§¸ ë ˆì´ì–´ë¥¼ ë³€ê²½í–ˆì„ ë•Œ, í•™ìŠµì— í° ì˜í–¥ì„ ì£¼ì—ˆê¸°ì— ì´ ë¶€ë¶„ì€ inception v3 ê¹Œì§€ ìœ ì§€ê°€ ëœë‹¤. ëª¨ë¸ì˜ ë§ˆì§€ë§‰ì—ëŠ” í”¼ì³ë§µì˜ í¬ê¸°ì™€ ê°™ì€ ì»¤ë„ ì‚¬ì´ì¦ˆë¡œ(ì´ ê²½ìš°ì—ëŠ” 7x7) Average Poolingì„ ìˆ˜í–‰í•˜ê³ , fully connected layerì™€ softmax í•¨ìˆ˜ë¡œ 1000ê°œì˜ í´ë˜ìŠ¤ì— ëŒ€í•œ ì˜ˆì¸¡ì„ ìˆ˜í–‰í•œë‹¤. íŒŒë¼ë¯¸í„° ìˆ˜ë¥¼ ëŠ˜ë¦¬ëŠ” ê°€ì¥ í° ìš”ì¸ ì¤‘ í•˜ë‚˜ê°€ fully connected layerì¸ë° ì´ë¥¼ ì ê²Œ ì‚¬ìš©í•¨ìœ¼ë¡œì¨ íŒŒë¼ë¯¸í„° ìˆ˜ë¥¼ ë” ì ˆì•½ í•  ìˆ˜ ìˆì—ˆë‹¤.

í‘œì—ëŠ” ë¬˜ì‚¬ë˜ì§€ ì•Šì•˜ì§€ë§Œ, ë…¼ë¬¸ì„ í™•ì¸í•´ ë³´ë©´ ì¤‘ê°„ì—ë„ ì˜ˆì¸¡í•˜ëŠ” ë¶€ë¶„ì´ ìˆë‹¤ëŠ” ê²ƒì„ ë°œê²¬ í•  ìˆ˜ ìˆë‹¤.(https://arxiv.org/pdf/1409.4842.pdf) ì´ ë¶€ë¶„ì„ "auxiliary classifiers"ë¼ê³  ë¶€ë¥¸ë‹¤. ê¹Šì€ ë§ì„ êµ¬ì„±í•œ íƒ“ì— gradientê°€ ëª¨ë“  ë ˆì´ì–´ë¡œ ì˜ í˜ëŸ¬ê°€ì§€ ì•Šì„ ìˆ˜ ìˆë‹¤ê³  ìƒê°í•˜ì—¬, ëª¨ë¸ ì¤‘ê°„ì— ì˜ˆì¸¡í•˜ëŠ” ë¶€ë¶„ì„ ì¶”ê°€í•˜ì—¬, ì—¬ê¸°ì—ì„œ ë‚˜ì˜¨ loss ê°’ì„ ìµœì¢… lossì— ë°˜ì˜í•˜ëŠ” íŠ¸ë¦­ì„ ì‚¬ìš©í–ˆë‹¤. ëª¨ë¸ ì¤‘ê°„ì—ì„œë„ ì˜ˆì¸¡ì„ í•¨ìœ¼ë¡œì¨, ì¼ì¢…ì˜ ì •ê·œí™” íš¨ê³¼ë¥¼ ê¸°ëŒ€í•œ ê²ƒì´ë‹¤. auxiliary classifier ì—ì„œ ê³„ì‚°ëœ loss ê°’ë“¤ì€ 0.3ì˜ ê°€ì¤‘ì¹˜ì™€ ê³±í•´ì ¸, ìµœì¢… loss ê°’ì— ë”í•´ì¡Œë‹¤. ê·¸ë¦¬ê³ , í…ŒìŠ¤íŠ¸ì‹œì—ëŠ” ì´ ë¶€ë¶„ì„ ì‚¬ìš©í•˜ì§€ ì•Šì•˜ë‹¤. GoogLeNetì—ì„œëŠ” ë‘ ê°œì˜ auxiliary classifierê°€ ì¡´ì¬í•œë‹¤.

## êµ¬í˜„

êµ¬í˜„í•¨ì— ìˆì–´, ì½”ë“œì˜ ê°„ê²°í•¨ì„ ìœ„í•´ auxiliary classifiers ë¶€ë¶„ì€ ìƒëµí•˜ì˜€ê³ , Local response normalizationì€ Batch normalizationìœ¼ë¡œ ëŒ€ì²´í•˜ì˜€ë‹¤.

```python
from keras import layers
from keras.datasets import mnist
from keras.models import Sequential, Model
from keras.optimizers import Adam
from keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_shape = (224, 224, 3)

input_ = layers.Input(shape=img_shape)
x = layers.Conv2D(64, 7, strides=2, padding='same')(input_)
x = layers.MaxPool2D(pool_size=(3, 3), strides=2, padding='same')(x)
x = layers.BatchNormalization()(x)  # ì •ê·œí™”. Local response normalization ëŒ€ì²´

x = layers.Conv2D(64, 1, strides=1)(x)
x = layers.Conv2D(192, 3, strides=1, padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(pool_size=(3, 3), strides=2, padding='same')(x)

x = inception_module(x, o_1=64, r_3=64, o_3=128, r_5=16, o_5=32, pool=32)
x = inception_module(x, o_1=128, r_3=128, o_3=192, r_5=32, o_5=96, pool=64)
x = layers.MaxPooling2D(pool_size=(3, 3), strides=2, padding='same')(x)

x = inception_module(x, o_1=192, r_3=96, o_3=208, r_5=16, o_5=48, pool=64)
x = inception_module(x, o_1=160, r_3=112, o_3=224, r_5=24, o_5=64, pool=64)
x = inception_module(x, o_1=128, r_3=128, o_3=256, r_5=24, o_5=64, pool=64)
x = inception_module(x, o_1=112, r_3=144, o_3=288, r_5=32, o_5=64, pool=64)
x = inception_module(x, o_1=256, r_3=160, o_3=320, r_5=32, o_5=128, pool=128)
x = layers.MaxPooling2D(pool_size=(3, 3), strides=2, padding='same')(x)

x = inception_module(x, o_1=256, r_3=160, o_3=320, r_5=32, o_5=128, pool=128)
x = inception_module(x, o_1=384, r_3=192, o_3=384, r_5=48, o_5=128, pool=128)
x = layers.AveragePooling2D(pool_size=(7, 7), strides=1)(x)

x = layers.Dropout(0.4)(x)
x = layers.Dense(1000)(x)
output = layers.Activation('softmax')(x)

googlenet = Model(input_, output)
```

### factorization

## Kerasì˜ Inception V3

```python
from keras.applications.inception_v3 import InceptionV3, decode_predictions

inceptionv3 = InceptionV3(input_shape=(299, 299, 3))
inceptionv3.summary()
```

```python
import cv2
import time

img = cv2.imread('bird1.jpg', -1)
img = cv2.resize(img, (299, 299))
img.shape
```

```python
start = time.time()
yhat = inceptionv3.predict(img.reshape(-1, 299, 299, 3))
time = time.time() - start
# label_key = np.argmax(yhat)
label = decode_predictions(yhat)
label = label[0][0]

print("í…ŒìŠ¤íŠ¸ ì‹œ ì†Œìš”ì‹œê°„: {}".format(time))
print("%s (%.2f%%)" % (label[1], label[2] * 100))
```
