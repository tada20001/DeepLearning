## Computational Graph

1. 노드 정의, 즉 그래프 만들기(Build graph using TensorFlow operations)
2. tf.Session()을 통한 세션 만들기 
3. 만든 세션을 통해 그래프 실행 : sess.run(op) : feed data and run graph
4. 그래프의 변수를 업데이트하거나 값을 반환

```
node1 = tf.constant(3.0, dtype=tf.float32)
node2 = tf.constant(4.0)
node3 = tf.add(node1, node2)
```

```
print("node1:", node1, "node2:", node2)
print("node3:", node3)
```

```
sess = tf.Session()
print("sess.run(node1, node2):", sess.run([node1, node2]))  # 2개 이상의 노드를 실행할 때는 리스트 형태로 입력해야 함
print("sess.run(node3):", sess.run(node3))
```

## Placeholder

그래프를 실행시키는 단계에서 값을 넘겨 주고 싶을 때는 

1. 노드를 tf.Placeholer( )로 만든다. 
2. 그래프를 실행시킬 때 feed_dict={} 로 값을 넘겨준다.

```
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)

adder_node = a + b

print(sess.run(adder_node, feed_dict={a: 3, b: 4.5}))
print(sess.run(adder_node, feed_dict={a: [1, 3], b: [2, 4]}))
```
