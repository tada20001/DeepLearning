### NN의 문제점
* layer를 많이 쌓으면 cost 함수 값과 accuracy가 일정 수준이 되면 떨어지지 않는 나쁜 결과를 낳는다.
* 그 이유는 Backprogation의 경우 2~3 layers에서는 잘 동작하지만 그 이상이 되면 잘 동작하지 않기 때문이다. $$$$

* **특히, 각 layer에서 미분한 값을 곱해서 최종 x, w의 영향력을 구했는데, 이 값들은 항상 sigmoid를 통과하는 값이기 때문에 항상 1보다 작고, 운이 좋지 않으면 0에 가까운 값이 된다. 이러한 값들이 각 레이어 수만큼 곱해지기 때문에 layer가 많아질수록 굉장히 작은 수치가 된다.**
$$$$
* 이 현상을 경사도가 사라진다고 한다 (Vanishing gradient).  즉, 앞단의 레이어들이 최종단에 영향을 미치지 않는다고 할 수 있다.

### sigmoid 사용에 대한 인식 전환

* 최대치를 1 이하로 한정시킨 sigmoid를 무한대로 놓는 ReLU(Rectified Linear Unit)를 적용하는 방법으로 위의 문제를 해결하게 된다.


```python
import tensorflow as tf
import numpy as np

tf.set_random_seed(777) 
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)
```

```python
X = tf.placeholder(tf.float32, shape=[None, 2], name='x')
Y = tf.placeholder(tf.float32, shape=[None, 1], name='y')

W1 = tf.Variable(tf.random_normal([2, 5]), name='weight1')  # input-layer
b1 = tf.Variable(tf.random_normal([5]), name='bias1')
L1 = tf.nn.relu(tf.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random_normal([5, 5]), name='weight2')
b2 = tf.Variable(tf.random_normal([5]), name='bias2')
L2 = tf.nn.relu(tf.matmul(L1, W2) + b2)
    
W3 = tf.Variable(tf.random_normal([5, 5]), name='weight3')
b3 = tf.Variable(tf.random_normal([5]), name='bias3')
L3 = tf.nn.relu(tf.matmul(L2, W3) + b3)

W4 = tf.Variable(tf.random_normal([5, 5]), name='weight4')
b4 = tf.Variable(tf.random_normal([5]), name='bias4')
L4 = tf.nn.relu(tf.matmul(L3, W4) + b4)
   
W5 = tf.Variable(tf.random_normal([5, 5]), name='weight5')
b5 = tf.Variable(tf.random_normal([5]), name='bias5')
L5 = tf.nn.relu(tf.matmul(L4, W5) + b5)
    
W6 = tf.Variable(tf.random_normal([5, 5]), name='weight6')
b6 = tf.Variable(tf.random_normal([5]), name='bias6')
L6 = tf.nn.relu(tf.matmul(L5, W6) + b6)

W7 = tf.Variable(tf.random_normal([5, 5]), name='weight7')
b7 = tf.Variable(tf.random_normal([5]), name='bias7')
L7 = tf.nn.relu(tf.matmul(L6, W7) + b7)

W8 = tf.Variable(tf.random_normal([5, 5]), name='weight8')
b8 = tf.Variable(tf.random_normal([5]), name='bias8')
L8 = tf.nn.relu(tf.matmul(L7, W8) + b8)

W9 = tf.Variable(tf.random_normal([5, 5]), name='weight9')
b9 = tf.Variable(tf.random_normal([5]), name='bias9')
L9 = tf.nn.relu(tf.matmul(L8, W9) + b9)

W10 = tf.Variable(tf.random_normal([5, 5]), name='weight10')
b10 = tf.Variable(tf.random_normal([5]), name='bias10')
L10 = tf.nn.relu(tf.matmul(L9, W10) + b10)
    
W11 = tf.Variable(tf.random_normal([5, 1]), name='weight11')  # output-layer
b11 = tf.Variable(tf.random_normal([1]), name='bias11')
hypothesis = tf.sigmoid(tf.matmul(L10, W11) + b11)
```

```python
# cost
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

# Accuracy computation
# True if hypothesis>0.5 else False
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

# Launch graph
with tf.Session() as sess:
    # tensorboard --logdir=./logs/xor_logs

    # Initialize TensorFlow variables
    sess.run(tf.global_variables_initializer())

    for step in range(10001):
        _, cost_val = sess.run([optimizer, cost], feed_dict={X: x_data, Y: y_data})

        if step % 100 == 0:
            print(step, cost_val)

    # Accuracy report
    h, p, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X: x_data, Y: y_data})
    
    print(f"\nHypothesis:\n{h} \nPredicted:\n{p} \nAccuracy:\n{a}")
```

### Actication Functions

* Sigmoid

* tanh : tanh(x)  -1 ~ 1

* ReLU  : max(0, x)

* Leaky ReLU : max(0.1x, x)

* Maxout : max(w1^T x + b1, w2^T x + b2)

* ELU

이 중 maxout, ReLU 순으로 성능이 좋다.
