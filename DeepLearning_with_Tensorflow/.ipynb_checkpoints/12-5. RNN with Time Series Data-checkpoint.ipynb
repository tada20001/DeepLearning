```python
import tensorflow as tf
import numpy as np
tf.logging.set_verbosity(tf.logging.ERROR)
```

### Many to One 형태의 RNN

* 예를 들어 7일 데이터에 대한 cell을 만들고 8일 값을 예측한다.

```python
input_dim = 5
sequence_length = 7 # 7일 데이터
output_dim = 1

xy = np.loadtxt("data-02-stock_daily.csv", delimiter=',')
xy[:5]
```

### Reading data

```python
xy = xy[::-1] # reverse order 

def MinMaxScaler(data):
    numerator = data - np.min(data, 0)
    denominator = np.max(data, 0) - np.min(data, 0)
    return numerator / (denominator + 1e-7)

xy = MinMaxScaler(xy)
x = xy
y = xy[:, [-1]] # close as label

dataX = []
dataY = []

for i in range(0, len(y) - sequence_length):
    _x = x[i:i+sequence_length]
    _y = y[i+sequence_length] # next close price
    print(_x,   "->", _y)
    dataX.append(_x)
    dataY.append(_y)
```

```python
len(dataX) # dataX.shape = (725, 7, 5)
```

```python
train_size = int(len(dataY) * 0.7)
test_size = len(dataY) - train_size
trainX, testX = np.array(dataX[:train_size]), np.array(dataX[train_size:])
trainY, testY = np.array(dataY[:train_size]), np.array(dataY[train_size:])
trainX.shape, testX.shape, trainY.shape, testY.shape
```

```python
tf.reset_default_graph()
hidden_size = 5
X = tf.placeholder(tf.float32, shape=[None, sequence_length, input_dim])
Y = tf.placeholder(tf.float32, shape=[None, output_dim])

# LSTM and Loss
cell = tf.contrib.rnn.BasicLSTMCell(num_units=hidden_size, state_is_tuple=True)
outputs, _states = tf.nn.dynamic_rnn(cell, X, dtype=tf.float32)
Y_pred = tf.contrib.layers.fully_connected(outputs[:, -1], output_dim, activation_fn=None)

# cost/ loss
loss = tf.reduce_sum(tf.square(Y_pred - Y))
# optimizer
optimizer = tf.train.AdamOptimizer(0.01).minimize(loss)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for i in range(1001):
    _, l = sess.run([optimizer, loss], feed_dict={X: trainX, Y:trainY})
    
    if i % 50 == 0:
        print("Step:", "%04d" % (i+1), "Loss:", l)
```

```python
testPredict = sess.run(Y_pred, feed_dict={X: testX})

import matplotlib.pylab as plt
%matplotlib inline
plt.plot(testY, label="True Y")
plt.plot(testPredict, label="Prediction")
plt.legend()
plt.show()
```
