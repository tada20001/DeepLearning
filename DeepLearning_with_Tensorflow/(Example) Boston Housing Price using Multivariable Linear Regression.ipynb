```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd

boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['const'] = np.ones(df.shape[0])
df.tail()
```

```python
import tensorflow as tf

# 1. 변수, placeholder 정의
n, m = df.shape

X = tf.placeholder(shape=[None, m], dtype=tf.float32)
Y = tf.placeholder(shape=[None, 1], dtype=tf.float32)

W = tf.Variable(tf.random_normal([m, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# 2. 선형 모델 만들기
hypothesis = tf.matmul(X, W) + b

# 3. 비용함수 정의
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# 4. 최소화 및 트레이닝 정의
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-6)
train = optimizer.minimize(cost)

# 5. 세션 런칭 및 실행
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for epoch in range(200):
    cost_val, _  = sess.run([cost, train], feed_dict={X: df.values, Y: boston.target.reshape([-1, 1])})
    if epoch % 10 == 0:
        print(epoch, "Cost:", cost_val)
```

Gradient descent algorithm으로는 cost가 줄어들지 않는다. (데이터 스케일 등 필요)

```python
X = tf.placeholder(shape=[None, m], dtype=tf.float32)
Y = tf.placeholder(shape=[None, 1], dtype=tf.float32)

XT = tf.transpose(X)
W = tf.matmul(tf.matmul(tf.matrix_inverse(tf.matmul(XT, X)), XT), Y)  # weight 값을 update 하지 않고 직접 계산한 연산에 대입

# model
hypothesis = tf.matmul(X, W)

with tf.Session() as sess:
    y_pred = sess.run(hypothesis, feed_dict={X: df.values, Y: boston.target.reshape(-1, 1)})
    
print("예측값:", y_pred[19], "실제 집값:", boston.target[19])
```
