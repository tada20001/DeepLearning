## autograd

autogradëŠ” PyTorchì—ì„œ í•µì‹¬ì ì¸ ê¸°ëŠ¥ì„ ë‹´ë‹¹í•˜ëŠ” í•˜ë¶€ íŒ¨í‚¤ì§€ì´ë‹¤. **autogradëŠ” í…ì„œì˜ ì—°ì‚°ì— ëŒ€í•´ ìë™ìœ¼ë¡œ ë¯¸ë¶„ê°’ì„ êµ¬í•´ì£¼ëŠ” ê¸°ëŠ¥ì„ í•œë‹¤.** í…ì„œ ìë£Œë¥¼ ìƒì„±í•  ë•Œ, **requires_grad** ì¸ìˆ˜ë¥¼ Trueë¡œ ì„¤ì •í•˜ê±°ë‚˜ **.requires_grad_(True)** ë¥¼ ì‹¤í–‰í•˜ë©´ ê·¸ í…ì„œì— í–‰í•´ì§€ëŠ” ëª¨ë“  ì—°ì‚°ì— ëŒ€í•œ ë¯¸ë¶„ê°’ì„ ê³„ì‚°í•œë‹¤. ê³„ì‚°ì„ ë©ˆì¶”ê³  ì‹¶ìœ¼ë©´ **.detach( )** í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ë©´ ëœë‹¤. 

ì˜ˆì œë¥¼ í†µí•´ ì•Œì•„ ë³´ë„ë¡ í•˜ì. requires_gradì¸ìˆ˜ë¥¼ Trueë¡œ ì„¤ì •í•˜ì—¬ Tensorë¥¼ ìƒì„±í–ˆë‹¤.

```python
# ë¯¸ë¶„
x = torch.rand(2, 2, requires_grad =True)
print(x)
```

ë‹¤ìŒìœ¼ë¡œ ì´ x ì— ì—°ì‚°ì„ ìˆ˜í–‰í•œë‹¤. ë‹¤ìŒ ì½”ë“œì˜ y ëŠ” ì—°ì‚°ì˜ ê²°ê³¼ì´ë¯€ë¡œ ë¯¸ë¶„ í•¨ìˆ˜ë¥¼ ê°€ì§„ë‹¤. **grad_fn** ì†ì„±ì„ ì¶œë ¥í•´ ë¯¸ë¶„ í•¨ìˆ˜ë¥¼ í™•ì¸í•  ìˆ˜ ìˆë‹¤.

```python
y = torch.sum(x * 3)
print(y, y.grad_fn)
```

**y.backward( )** í•¨ìˆ˜ë¥¼ ì‹¤í–‰í•˜ë©´ x ì˜ ë¯¸ë¶„ê°’ì´ ìë™ìœ¼ë¡œ ê°±ì‹ ëœë‹¤. xì˜ **grad** ì†ì„±ì„ í™•ì¸í•˜ë©´ ë¯¸ë¶„ê°’ì´ ë“¤ì–´ ìˆëŠ” ê²ƒì„ í™•ì¸ í•  ìˆ˜ ìˆë‹¤. y ë¥¼ êµ¬í•˜ê¸° ìœ„í•œ x ì˜ ì—°ì‚°ì„ ìˆ˜ì‹ìœ¼ë¡œ ì“°ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤.

$$y = \displaystyle\sum_{i=1}^4 3 \times x_i$$

ì´ë¥¼  ğ‘¥_ğ‘– ì— ëŒ€í•´ ë¯¸ë¶„ í•˜ë©´ ë¯¸ë¶„ í•¨ìˆ˜ëŠ” ë‹¤ìŒê³¼ ê°™ë‹¤.

$$\dfrac{\partial y}{\partial x_i} = 3$$

ì‹¤ì œ ë¯¸ë¶„ê°’ê³¼ ê°™ì€ì§€ í™•ì¸í•´ ë³´ì.

```python
print(x.grad)
```

```python
y.backward(retain_graph=True)
```

```python
x.grad
```

**backward()í•¨ìˆ˜ëŠ” ìë™ìœ¼ë¡œ ë¯¸ë¶„ê°’ì„ ê³„ì‚°í•´ requires_grad ì¸ìˆ˜ê°€ Trueë¡œ ì„¤ì •ëœ ë³€ìˆ˜ì˜ grad ì†ì„±ì˜ ê°’ì„ ê°±ì‹ í•œë‹¤.** retain_graph ë¯¸ë¶„ì„ ì—°ì‚°í•˜ê¸° ìœ„í•´ì„œ ì‚¬ìš©í–ˆë˜ ì„ì‹œ ê·¸ë˜í”„ë¥¼ ìœ ì§€í•  ê²ƒì¸ê°€ë¥¼ ì„¤ì •í•˜ëŠ” ê²ƒì´ë‹¤. 

ê¸°ë³¸ê°’ì€ Falseë¡œ ì„¤ì •ë˜ì–´ ìˆì§€ë§Œ ë™ì¼í•œ ì—°ì‚°ì— ëŒ€í•´ ì—¬ëŸ¬ ë²ˆ ë¯¸ë¶„ì„ ê³„ì‚°í•˜ê¸° ìœ„í•´ì„œëŠ” Trueë¡œ ì„¤ì •ë˜ì–´ ìˆì–´ì•¼ í•œë‹¤. ê·¸ë¦¬ê³  **ë¯¸ë¶„ê°’ì„ ê·¸ëŒ€ë¡œ ì¶œë ¥ë°›ì•„ ì‚¬ìš©í•˜ê³  ì‹¶ì€ ê²½ìš°ì—ëŠ” torch.autograd.grad( )í•¨ìˆ˜ì— ì¶œë ¥ê°’ê³¼ ì…ë ¥ê°’ì„ ì…ë ¥í•˜ë©´ ë¯¸ë¶„ê°’ì„ ì¶œë ¥í•œë‹¤.**

## PyTorchë¥¼ ì´ìš©í•œ í¼ì…‰íŠ¸ë¡  êµ¬í˜„

ì´ë²ˆì—ëŠ” ë¶“ê½ƒ ë¶„ë¥˜ ë¬¸ì œë¥¼ í•´ê²°í•˜ëŠ” í¼ì…‰íŠ¸ë¡  ëª¨í˜•ì„ êµ¬í˜„í•œë‹¤.

```python
from sklearn.datasets import load_iris

iris = load_iris()

idx = np.in1d(iris.target, [0, 2]) # label 0, 2 ì‚¬ìš©
x = torch.from_numpy(iris.data[idx, :2]).type(torch.float64)
y = torch.from_numpy((iris.target[idx] - 1.0)[:, np.newaxis]).type(torch.float64) # -1, 1ë¡œ ë³€í™˜
```

ëª¨í˜•ì— ì‚¬ìš©í•˜ëŠ” ë³€ìˆ˜ë¥¼ **requires_grad** ì¸ìˆ˜ë¥¼ Trueë¡œ í•˜ì—¬ ìƒì„±í•œë‹¤.

```python
w = torch.rand((2, 1), dtype=torch.float64, requires_grad=True)
b = torch.rand((1, 1), dtype=torch.float64, requires_grad=True)
```

ìƒì„±í•œ ë³€ìˆ˜ì— ì§ì ‘ ì—°ì‚°ì„ ìˆ˜í–‰í•´ì„œëŠ” ì•ˆëœë‹¤. ì˜ˆì™€ ê°™ì´ "w" ë¼ëŠ” ë³€ìˆ˜ëª…ìœ¼ë¡œ ë³€ìˆ˜ë¥¼ í• ë‹¹í•˜ê³  ì´ ë³€ìˆ˜ë¥¼ ì—°ì‚°í•  ë•Œ íŒŒì´ì¬ì—ì„œ í•˜ë“¯ì´

ì´ë ‡ê²Œ í•´ì„œëŠ” ì•ˆëœë‹¤. ë³€ìˆ˜ì— ì—°ì‚°ì˜ ê²°ê³¼ ê°’ì´ ë“¤ì–´ê°€ë©´ ì´í›„ í•´ë‹¹ ë³€ìˆ˜ëŠ” ì´ì „ì— ìƒì„±í•œ ë³€ìˆ˜ê°€ ì•„ë‹ˆê¸° ë•Œë¬¸ì— ë” ì´ìƒ requires_grad ê°€ Trueë¡œ ì„¤ì •ë˜ì§€ ì•ŠëŠ”ë‹¤. **ë³€ìˆ˜ì— ì—°ì‚°, í• ë‹¹ì„ í•˜ê³  ì‹¶ì„ ë•ŒëŠ” ë³€ìˆ˜ì˜ data ì†ì„±ì„ ë¶ˆëŸ¬ì™€ ë³€ê²½í•œë‹¤.**

```python
w.data *= 0.001
b.data *= 0.001
```

### ìµœì í™”

ë‹¤ìŒìœ¼ë¡œ SGD ìµœì í™”ë¥¼ ìˆ˜í–‰í•œë‹¤. **.backward( )**ë¥¼ ì‚¬ìš©í•˜ë©´ ë¯¸ë¶„ê°’ì´ ê³„ì† ëˆ„ì ëœë‹¤. ìš°ë¦¬ëŠ” í•™ìŠµí•œ ë’¤ ê°€ì¤‘ì¹˜ë¥¼ ìˆ˜ì •í•˜ê³  ìˆ˜ì •í•œ ê°€ì¤‘ì¹˜ì— ëŒ€í•œ ë¯¸ë¶„ê°’ìœ¼ë¡œ ë‹¤ì‹œ í•™ìŠµí•˜ëŠ” ê²ƒì´ê¸° ë•Œë¬¸ì— ê°€ì¤‘ì¹˜ë¥¼ ìˆ˜ì •í•œ ë’¤ì— **.grad_zero_( )** í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•˜ì—¬ ë³€ìˆ˜ì— í• ë‹¹ëœ ë¯¸ë¶„ê°’ì„ 0ìœ¼ë¡œ ë§Œë“¤ì–´ ì¤€ë‹¤.

```python
learning_rate = 0.001
epochs = 300

for epoch in range(epochs):
    z = (x.mm(w) + b).tanh()
    loss_vec = torch.mul(-y, z)
    loss = torch.sum(torch.max(torch.cat([torch.zeros_like(loss_vec), loss_vec], dim=-1), dim=-1)[0], dim=-1)
    loss.backward()
    w.data -= w.grad * learning_rate
    b.data -= b.grad * learning_rate
    
    if epoch % 40 == 0:
        print("{:3} - loss : {}".format(epoch, loss.item()))
        
    w.grad.zero_()
    b.grad.zero_()
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y, [-1 if p < 0 else 1 for p in z.detach().numpy()])
```

## PyTorchë¥¼ ì´ìš©í•œ ì‹ ê²½ë§ êµ¬í˜„

PyTorch ë‹¤ì–‘í•œ ì‹ ê²½ë§ ëª¨í˜•ì„ ì‰½ê²Œ êµ¬í˜„ í•  ìˆ˜ ìˆë„ë¡ **torch.nn.module** ì„ ì œê³µí•œë‹¤. ì´ë¥¼ ì´ìš©í•´ ê°„ë‹¨í•œ ì‹ ê²½ë§ ëª¨í˜•ì„ êµ¬í˜„í•œë‹¤. PyTorchì—ì„œëŠ” ìƒ˜í”Œ ë°ì´í„°ë¥¼ ì œê³µí•˜ì§€ ì•Šìœ¼ë¯€ë¡œ ìƒ˜í”Œë°ì´í„°ëŠ” Kerasë¥¼ ì´ìš©í•´ MNIST ìˆ«ì ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¤ê² ë‹¤.

ì´ì „ì— Kerasì—ì„œ í–ˆë˜ ë°©ì‹ê³¼ ê°™ì´ ì „ì²˜ë¦¬ë¥¼ ì§„í–‰í•œë‹¤. ë‹¤ë§Œ ì—¬ê¸°ì„œëŠ” ì •ë‹µ ë°ì´í„°ì— ëŒ€í•œ ì›í•« ì¸ì½”ë”©ì‘ì—…ì„ í•˜ì§€ ì•ŠëŠ”ë‹¤. ì´ìœ ëŠ” ë°”ë¡œ ë‹¤ìŒì— ì„¤ëª…í•  ê²ƒì´ë‹¤.

ì „ì²˜ë¦¬ë¥¼ ìˆ˜í–‰í–ˆìœ¼ë©´ PyTorchê°€ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” í…ì„œë¡œ ë³€í™˜í•˜ê³  ê°ê° ì•Œë§ëŠ” ë°ì´í„° íƒ€ì…ìœ¼ë¡œ ë³€í™˜í•œë‹¤. ì´ë¯¸ì§€ ë°ì´í„°ì— ëŒ€í•´ì„œëŠ” float íƒ€ì…, ì¶œë ¥ë°ì´í„°ì— ëŒ€í•´ì„œëŠ” int64 íƒ€ì…ìœ¼ë¡œ ë³€í™˜í•œë‹¤. ì´ëŠ” ì¶”í›„ ì‚¬ìš©í•  PyTorchì˜ í•¨ìˆ˜ê°€ í•´ë‹¹ ë°ì´í„° íƒ€ì…ì„ ì§€ì›í•˜ê¸° ë•Œë¬¸ì´ë‹¤.

```python
X_train = X_train0.reshape(-1, 28*28).astype('float64') / 255.0
X_test = X_test0.reshape(-1, 28*28).astype('float64') / 255.0
```

```python
X_train, y_train, X_test, y_test = map(torch.tensor, (X_train, y_train, X_test, y_test))

X_train, y_train, X_test, y_test = X_train.float(), y_train.long(), X_test.float(), y_test.long()
```

**torch.nn.Sequential** ì„ ì´ìš©í•˜ë©´ ì†ì‰½ê²Œ ëª¨ë¸ì„ êµ¬ì„±í•  ìˆ˜ ìˆë‹¤. ì´ì „ Kerasì˜ ë°©ì‹ê³¼ ìœ ì‚¬í•˜ë‹¤. ë‹¤ë§Œ ëª…ë ¹ì–´ê°€ ì¡°ê¸ˆì”© ë‹¤ë¥¼ ë¿ì´ë‹¤. 

Kerasë¡œ êµ¬í˜„í•œ ê²ƒê³¼ ë™ì¼í•œ êµ¬ì¡°ë¡œ êµ¬ì„±í•œë‹¤. **torch.nn.Linear**ëŠ” ë ˆì´ì–´ ê°„ì˜ ì„ í˜•ê²°í•©ì„ ì˜ë¯¸í•˜ê³  ì´ë¥¼ í™œì„±í™”í•¨ìˆ˜ì— í†µê³¼ì‹œí‚¤ë©´ ì´ì „ì— ë°°ì› ë˜ **MLP(multi layer perceptron)** êµ¬ì¡°ê°€ ëœë‹¤. 28 x 28 í¬ê¸°ì˜ ì´ë¯¸ì§€ë¥¼ 1ì°¨ì› ë²¡í„°ë¡œ ë§Œë“¤ì–´ ì…ë ¥í•˜ê³  15ê°œ ë…¸ë“œì˜ ì€ë‹‰ì¸µì„ ê±°ì³ 10ê°œ ë…¸ë“œë¥¼ ì¶œë ¥í•œë‹¤. ì´ ëª¨í˜•ì„ GPUë¥¼ ì‚¬ìš©í•´ í•™ìŠµì‹œí‚¤ê³  ì‹¶ë‹¤ë©´ ëª¨í˜•ê³¼ ì‚¬ìš©í•  ë°ì´í„°ë¥¼ GPUì—°ì‚° ê°€ëŠ¥ ìë£Œí˜•ìœ¼ë¡œ ë³€í™˜í•˜ë©´ ëœë‹¤.

```python
## create model
from torch import nn

model = nn.Sequential(
    nn.Linear(784, 15),  # hidden layer 15ê°œ
    nn.Sigmoid(),       # í™œì„±í™” í•¨ìˆ˜
    nn.Linear(15, 10),  # ì¶œë ¥ shape
    nn.Sigmoid(),
)
```

PyTorchì—ì„œ ìµœì í™” ì•Œê³ ë¦¬ì¦˜ì€ **torch.optim**ì— ì •ì˜ë˜ì–´ ìˆë‹¤. ë‹¤ìŒì²˜ëŸ¼ torch.optim í´ë˜ìŠ¤ì—ì„œ ì›í•˜ëŠ” ìµœì í™” ë°©ë²•ì„ ì •ì˜í•œë‹¤. 

ì—¬ê¸°ì„œëŠ” SGDë°©ë²•ì„ ì‚¬ìš©í•œë‹¤. ê·¸ë¦¬ê³  ëŒ€ë¶€ë¶„ì˜ ë¹„ìš©í•¨ìˆ˜ëŠ” torch.nn í´ë˜ìŠ¤ì— êµ¬í˜„ë˜ì–´ ìˆë‹¤. ì—¬ê¸°ì„œëŠ” í¬ë¡œìŠ¤ ì—”íŠ¸ë¡œí”¼ ë¹„ìš©í•¨ìˆ˜(Cross Entropy Loss)ë¥¼ ì‚¬ìš©í•œë‹¤. í¬ë¡œìŠ¤ ì—”íŠ¸ë¡œí”¼ ë¹„ìš©í•¨ìˆ˜ëŠ” ì‹ ê²½ë§ ëª¨í˜•ì„ ì´ìš©í•´ ë¶„ë¥˜ë¬¸ì œë¥¼ í’€ ë•Œ ë§ì´ ì‚¬ìš©í•˜ëŠ” ë¹„ìš©í•¨ìˆ˜ì´ë‹¤. ì´ëŠ” ì¶”í›„ ì‹ ê²½ë§ ì„±ëŠ¥ê°œì„  ë‹¤ë£° ê²ƒì´ë‹¤.

```python
## cost function and optimizer
from torch import optim

opt = optim.SGD(model.parameters(), lr=0.5)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
def accuracy(pred, y):
    preds = torch.argmax(pred, dim=1)
    return (preds == y).float().mean()

# training the model
epochs = 10
n = X_train.shape[0]
batch_size = 100

for epoch in range(epochs):
    for i in range(int(n / batch_size)):
        start_i = i * batch_size
        end_i = start_i + batch_size
        xb = X_train[start_i:end_i]
        yb = y_train[start_i:end_i]
        pred = model(xb)
        loss = loss_fn(pred, yb)
        loss.backward()
        opt.step()
        opt.zero_grad()
        
    acc = accuracy(pred, yb)
    print("{} epoch, cross entropy : {}, accuracy : {}".format(epoch + 1, loss.detach().item(), acc.detach().item()))
```
