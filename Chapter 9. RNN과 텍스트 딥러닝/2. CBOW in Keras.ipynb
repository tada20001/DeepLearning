### CBOW (continuos Bag of Words) Embedding

하나의 단어로부터 다음에 오는 단어를 예측하는 문제. 여러 개의 단어를 나열한 뒤, 이아 관련된 단어를 추정하는 문제이다. n개의 단어 열로부터 다음 단어를 예측하는 문제가 된다. 

```python
import numpy as np

import keras.backend as K
from keras.models import Sequential
from keras.layers import *
from keras.utils import np_utils
from keras.utils.vis_utils import model_to_dot
from keras.preprocessing import sequence
from keras.preprocessing.text import Tokenizer
```

```python
import gensim
import nltk

from IPython.display import SVG
```

```python
nltk.download("gutenberg")
nltk.download('punkt')
```

```python
sentences = [" ".join(s) for s in nltk.corpus.gutenberg.sents("carroll-alice.txt") if len(s) > 2] # 3개이상 단어의 문장 추출
sentences[:5]
```

```python
len(sentences)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
corpus = tokenizer.texts_to_sequences(sentences)
len(corpus)
```

```python
corpus[0]
```

```python
nb_samples = sum(len(s) for s in corpus)  # 전체 문장에 쓰인 단어 수
V = len(tokenizer.word_index) + 1 # 단어 수 + 1
dim = 100
window_size = 2
```

```python
tokenizer.word_index
```

```python
def generate_data(corpus, window_size, V):
    maxlen = window_size * 2
    for words in corpus:
        L = len(words) # 각 문장별 단어 수
        for index, word in enumerate(words):
            contexts = []
            labels = []
            s = index - window_size
            e = index + window_size + 1
            
            contexts.append([words[i] for i in range(s, e) if 0 <= i < L and i != index])
            labels.append(word)  # 단어 번호 추가
            
            x = sequence.pad_sequences(contexts, maxlen=maxlen)
            y = np_utils.to_categorical(labels, V)  # one_hot encoding
            yield (x, y)
```

```python
X = []
Y = []
for x, y in generate_data(corpus, window_size, V):
    X.append(x)
    Y.append(y)
    
X[:5]
```

```python
Y[:5]
```

```python
X = np.concatenate(X)
Y = np.concatenate(Y)
X[:5]
```

## Model

```python
V, dim
```

```python
cbow = Sequential()
cbow.add(Embedding(input_dim=V, output_dim=dim, input_length=window_size*2)) # input_dim=2572, output_dim=100
cbow.add(Lambda(lambda x: K.mean(x, axis=1), output_shape=(dim,)))
cbow.add(Dense(V, activation='softmax'))
cbow.compile(loss='categorical_crossentropy', optimizer='adadelta', metrics=['accuracy'])
```

```python
%%time
hist = cbow.fit(X, Y, epochs=1000, batch_size=300, verbose=2)
```

## Word2Vec

```python
vectors = cbow.get_weights()[0]
vectors.shape
```

```python
f = open('cbow_vectors.txt', 'w')
f.write('{} {}\n'.format(V-1, dim))
for word, i in tokenizer.word_index.items():
    str_vec = ' '.join(map(str, list(vectors[i, :]))) 
    f.write('{} {}\n'.format(word, str_vec)) # 각 단어별 가중치 리스트 생성 
f.close()
```
